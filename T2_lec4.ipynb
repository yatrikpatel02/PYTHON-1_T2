# Important functions of tuple

```python
# 1) len()
t=(20,30,40,50)
print(len(t))
```

```python
# 2) count()
t=(1,2,2,3,3,3,1,1,2,4,5)
print(t.count(2))
print(t.count(1))
print(t.count(27))
```

```python
# 3) index()
# returnns the index first occurance of given element, if specified element is not available then we will get valueError
t=(10,20,10,1,0,30)
print(t.index(10))
print(t.index(10,2,4))
#print(t.index(8))       #gives valueError
```

```python
# index function also work for string
s="Hello world"
print(s.index("l"))
```

```python
# 4) sorted()
t=(10,30,40,50,20)
print(sorted(t))
```

```python
# sorted also work for string
x=""
s="Hello"
s1=sorted(s)
print(s1)
for i in s:
    x=x+i
    print(x)
```

```python
t=(10,20,40,30,50)
t1=sorted(t,reverse=True)
print(t1)
```

```python
# 5) min() and max() fumction
t=(10,20,40,50,30)
print(min(t))
print(max(t))
```

```python
# tuple packing and unpacking
a=10
b=20
c=30
d=40
t=a,b,c,d
print(t)
```

```python
t=(10,20,30,40)
a,b,c,d=t
print(a,b,c,d)
```

```python
# loop through tuple
t=("apple","banana","cherry")
for i in t:
    print(t)
```

```python
t=("apple","banana","cherry")
for i in range(len(t)):
    print(t[i])
```

```python
t=("apple","banana","cherry")
i=0
while i!=len(t):
    print(t[i])
    i=i+1
```

```python
# 6) reversed()
# compute the reverse of the given sequence object and returns it in form of list
s="python"
print(list(reversed(s)))
t=(30,20,40,50,10)
print(list(reversed(t)))
r=range(5,9)
print(list(reversed(r)))
l=[1,2,3,4,5]
print(list(reversed(l)))
```

# Accessing the element of list

```python
# 1) by using index()
l=[1,2,3,4,5]
print(l[0])
print(l[2])
# print(l[5])    # gives error
```

```python
# by using slicing operation
l=[1,2,3,4,5,6,7,8,9]
print(l[2:7:2])
print(l[4::2])
print(l[::-1])
print(l[3:7])
print(l[8:2:-2])
print(l[9:0:-1])
```

```python
# mathematical operators
# 1) concatenation (+)
a=[1,2,3]
b=[4,5,6]
c=a+b
print(c)
d=a+[4]
print(d)
```

```python
# 2) repetition (*)
l=[1,2,3]
x=l*2
print(x)
```

```python
# membership operator
l=[10,20,30,40,50]
print(10 in l)
print(10 not in l)
print(50 in l)
print(60 not in l)
```
