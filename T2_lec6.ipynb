```python
# given a list of size L ypu need to count the no of special elememts in the given list an element is special if removal of that element make the list balance, the list will be balanced if sum of the even index element is equal to sum of odd index elements, also print the updated list after removal of special element
l=[5,5,2,5,8]
l1=l.copy()
special=0
print(l1)
for j in range(len(l)):
    odd=0
    even=0
    l1.pop(j)
    for i in range(len(l1)):
        if i%2!=0:
            odd=odd+l1[i]
        else:
            even=even+l1[i]
    if odd==even:
        special+=1
        print("special number is : ",l[j]," of index : ",j)
    l1=l.copy()
print("Total special no are: ",special)    
```

```python
# wap to count the no of string where the string length is 3 or more and the first and last character are same from the given list of string
l=["abc","XYZ","aba","2112","128451","ab","12345"]
count=0
for i in range(len(l)):
    if len(l[i])>=3:
        s=l[i]
        if s[0]==s[len(s)-1]:
            count+=1
print("total no of string with the given case = ",count)            
```

```python
l=[[1,2,3],[4,5,6],[7,8,9]]
print("original matrix - ")
for i in l:
    print(i)
print()    
print("transpose matrix - ")    
for i in range(len(l)):
    print(end="[")
    for j in range(len(l[i])):
        print("",l[j][i],end=", ")
    print(end="]")
    print()
```

# dictionary

###### represent group of objects as key value pairs
###### duplicate keys are not allowed nut values can be duplicate
###### heterogeneous objects alllowed for both key and value
###### insertion order not preserved
###### indexing and slicing concept are not applicable

```python
d={10:"lucky",20:"arman",30:"dhairya",10:"aryan"}
# creation of dictionary
d={}
# or d=dict() 
d[11]="lucky"
d[20]="arman"
d[30]="dhairya"
print(d)
```

##### updating  the elements of dictionary

```python
d={10:"lucky",20:"arman",30:"dhairya",10:"aryan"}
d[40]="sumit"
print(d)
d[10]="jassu"
print(d)
```

##### deleting the elements from dictionary


```python
d={10:"lucky",20:"arman",30:"dhairya",40:"abstr",50:"chaas"}
del d[10]
print(d)
del d[20]
print(d)
# del d # del delete the whole dictionary
# clear do not delete the dictionary
d.clear()
print(d)
```
