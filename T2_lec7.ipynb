# important functions of set

```python
# 1) len()
# 2) add(x) - add item x to the set
s={10,20,30}
s.add(40)
print(s)
print(len(s))
```

```python
# 3) update() - update(x,y,z)
s={10,20,30}
l=[10,20,30,40]
s.update(l,range(5))
print(s)
```

```python
# 4) copy()
s={10,20,30}
s1=s.copy()
print(s)
print(s1)
s1.add(40)
print(s)
print(s1)
```

```python
# 5) pop()
s={10,20,30,40}
print(s.pop())
print(s)
x={'a','b','c','d'}
print(x.pop())
print(x)
```

```python
# 6) remove(x)
s={10,20,30,40}
s.remove(10)
print(s)
# s.remove(1)     # if element not present then it will give error

# 7) discard(x) # if not present then no error
s.discard(20)
print(s)
s.discard(1)
print(s)
```

```python
# 8) clear()
s={10,20,30,40}
s.clear()
print(s)
```

```python
# set comprehension
s={x*x for x in range(5)}
print(s)
s1={2**x for x in range(2,10,2)}
print(s1)
```

```python
# frozenset()
# makes set immutable
s={'apple','banana','cherry'}
x=frozenset(s)
x.add("orange")
print(x)
```

# lambda function
##### function without name
##### nameless function is called anonymous function or lambda function

```python
def square(n):
    return n*n
print(square(7))
print(square(5))

s=lambda n:n*n
print(s(5))
```

```python
s=lambda a,b:a+b
print(s(10,20))
s1=lambda a,b:(a+b,a-b,a*b,a//b)
print(s1(10,20))
```

```python
s=lambda x:x.upper()
print(s("geet"))
```

```python
words={'apple','banana','cherry','kiwi'}
s=sorted(words,key=lambda x:len(x))
print(s)
```

```python
s=[(1,3),(4,2),(5,8),(9,1)]
s.sort(key=lambda x:x[1])
print(s)
```

```python
# map() - for every element present in given sequence apply same functionality and generate new element with required modification
# synrax - map(function,sequence)
l=[1,2,3,4,5]
def double(x):
    return 2*x
l1=list(map(double,l))
print(l1)
```

```python
l=[1,2,3,4,5]
l1=tuple(map(lambda x:x*2,l))
print(l1)
```

```python
l1=[1,2,3,4,5]
l2=[2,3,4,5,6]
l3=list(map(lambda x,y:x+y,l1,l2))
print(l3)
```

```python
# reduce() - reduces sequence of elements into a single element by applying specific function
# syntax - reduce(function_name,sequence)
# reduce() function is present in functools module so we need to import the module before using it
from functools import *
l=[10,20,30,40,50]
sum=reduce(lambda x,y:x+y,l)
print(sum)
mul=reduce(lambda x,y:x*y,l)
print(mul)
```

```python
# filter() - used to filter values from given sequence based on same condition
# syntax - filter(function,sequence)
# function argument is responsible to perform conditional check on sequence
def even(x):
    if x%2==0:
        return True
    else:
        return False
l=[0,5,10,15,20,25,30]
l1=list(filter(even,l))
print(l1)
l2=list(filter(lambda x:x%2==0,l))
print(l2)
```

```python
check=lambda x:"Even" if x%2==0 else "odd"
print(check(5))
print(check(4))
```

```python
#wap using lambda function with map to capitalize the first letter of each word in a list
l=['apple','banana','cherry']
l1=list(map(lambda x:x.capitalize(),l))
print(l1)
```

```python
#wap to calculate the sum of positive nd negative numbers from the given list of numbers using lambda function
l=[1,2,-1,10,-100,34,-7]
pos=list(filter(lambda x:x>=0,l))
neg=list(filter(lambda x:x<0,l))
print(pos)
print(neg)
pos_sum=reduce(lambda x,y:x+y,pos)
neg_sum=reduce(lambda x,y:x+y,neg)
print(pos_sum)
print(neg_sum)
```
